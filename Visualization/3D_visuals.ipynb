```python
#In this file, I continue to visualize the normalized position player data,
#creating additional 3D representations using Python's matplotlib
```

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
non_rookies = pd.read_csv('/content/drive/MyDrive/Data_Science_Projects/positionplayers_cleaned.csv')
```

```python
non_rookies.drop(columns = ['Unnamed: 0'], inplace = True)
non_rookies.head()
```

```python
#Filter out the MLB position players on rookie contracts for visualization purposes
non_rookies = non_rookies.loc[non_rookies['Rookie contract'] == 0]
non_rookies.shape
```

# 3D Scatter Plots

```python
#In the code below I use various combinations of features to see how both relate to player salary.
```

```python
import matplotlib.pyplot as plt
x = non_rookies['ops']
y = non_rookies['WAR']
z = non_rookies['salary']

fig = plt.figure(figsize = (12, 10))
ax = plt.axes(projection ="3d")

ax.scatter3D(x, y, z, color = 'blue', alpha = 0.6)
plt.title('Salary vs. WAR vs. OPS')
ax.set_xlabel('OPS', fontweight ='bold')
ax.set_ylabel('WAR', fontweight ='bold')
ax.set_zlabel('Player Salary', fontweight ='bold')
```
