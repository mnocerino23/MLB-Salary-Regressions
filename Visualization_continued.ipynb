```python
import matplotlib.pyplot as plt
x = stats_normalized['ops']
y = stats_normalized['WAR']
z = stats_normalized['salary']

fig = plt.figure(figsize = (12, 10))
ax = plt.axes(projection ="3d")

ax.scatter3D(x, y, z, color = 'blue', alpha = 0.6)
plt.title('Salary vs. WAR vs. OPS')
ax.set_xlabel('OPS', fontweight ='bold')
ax.set_ylabel('WAR', fontweight ='bold')
ax.set_zlabel('Player Salary', fontweight ='bold')
```
