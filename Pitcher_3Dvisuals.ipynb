```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nonrookie_pitchers = pitchers.loc[pitchers['age'] > 25]
nonrookie_relievers = pitchers.loc[(pitchers['age'] > 25) & (pitchers['POS1'] == 'Reliever')] 
nonrookie_starters = pitchers.loc[(pitchers['age'] > 25) & (pitchers['POS1'] == 'Starter')]
```

```python
print('Non-rookie Pitchers Shape', nonrookie_pitchers.shape)
print('Non-rookie Relivers Shape', nonrookie_relievers.shape)
print('Non-rookie Starters Shape', nonrookie_starters.shape)
```

# Age

```python
sns.scatterplot(data=nonrookie_relievers, x="age", y="salary",alpha = 0.5).set(title = 'Salary vs. Age (Relievers)')
```

```python
sns.scatterplot(data=nonrookie_starters, x="age", y="salary", alpha = 0.5).set(title='Salary vs. Age (Starters)')
```
